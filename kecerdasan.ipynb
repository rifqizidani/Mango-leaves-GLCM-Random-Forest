```python
#ANTHRACNOSE

import os
import cv2
files = os.listdir('/tmp/daun-harumanis-kagle/Anthracnose')

for i in range(file_count):
  img = cv2.imread('/tmp/daun-harumanis-kagle/Anthracnose/' + files[i])

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  glcm = greycomatrix(gray,
                    distances=[1],
                    angles=[0],
                    levels=gray.max()+1,
                    symmetric=True)
  
  contrast_img = greycoprops(glcm, 'contrast')[0]
  correlation_img = greycoprops(glcm, 'correlation')[0]
  energy_img = greycoprops(glcm, 'energy')[0]
  homogeneity_img = greycoprops(glcm, 'homogeneity')[0]
  dissimilarity_img = greycoprops(glcm, 'dissimilarity')[0]
  asm_img = greycoprops(glcm, 'ASM')[0]
  kondisi_img = 'Anthracnose'

  with open('/content/drive/MyDrive/AI/ai.csv', mode='a') as csv_file:
    # membuat objek writer
    writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    # menulis baris ke file CSV
    writer.writerow([files[i],contrast_img[0],correlation_img[0],energy_img[0],homogeneity_img[0],dissimilarity_img[0],asm_img[0],kondisi_img])
```
